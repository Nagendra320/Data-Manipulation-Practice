**1. Advanced Stock Market Analysis**

**Sample DataFrame**:
```python
import pandas as pd

stock_data = pd.DataFrame({
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Company_A': [100, 102, 101, 99, 98],
    'Company_B': [110, 108, 109, 111, 112],
    'Company_C': [120, 118, 119, 121, 122],
    'Volume_A': [2000, 2500, 2400, 2300, 2200],
    'Volume_B': [2100, 2600, 2500, 2400, 2300],
    'Volume_C': [2200, 2700, 2600, 2500, 2400]
})
```

**Given Query**: 
"Calculate the percentage change in stock prices for each company between the first and the last day. Next, identify which company had the highest average trading volume. Finally, create a new DataFrame showing the day-wise total trading volume across all companies."


````python
stock_data = pd.DataFrame({
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Company_A': [100, 102, 101, 99, 98],
    'Company_B': [110, 108, 109, 111, 112],
    'Company_C': [120, 118, 119, 121, 122],
    'Volume_A': [2000, 2500, 2400, 2300, 2200],
    'Volume_B': [2100, 2600, 2500, 2400, 2300],
    'Volume_C': [2200, 2700, 2600, 2500, 2400]
})
````

````python
stock_data
````

````python
#The percentage change in stock prices for each company between the first and the last day.
price_change=stock_data.set_index('Date',drop=True)[['Company_A','Company_B','Company_C']].pct_change(periods= stock_data.shape[0]-1).iloc[-1]*100

#Company with highest average trading volume.
High_trading_volume=stock_data[['Volume_A','Volume_B','Volume_C']].mean().idxmax().split('_')[1]

#New column with total trading volume
stock_data['Total_trading_volume']=stock_data['Volume_A']+stock_data['Volume_B']+stock_data['Volume_C']

print(f'Percentage change in stock prices:')
print(price_change)
print(f'\nCompany with the highest average trading volume: Company {High_trading_volume}')
print(f'\nDay-wise total trading volume across all companies:')
print(stock_data[['Date','Total_trading_volume']])
````


**2. Comprehensive Student Exam Analysis**

**Sample DataFrame**:
```python
student_data = pd.DataFrame({
    'Student_ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Math': [92, 85, 88, 76, 95],
    'Physics': [85, 90, 82, 89, 92],
    'Chemistry': [88, 85, 91, 87, 93],
    'Biology': [90, 88, 94, 86, 89]
})
```

**Given Query**: 
"First, calculate the average score for each student across all subjects. Next, identify the top scorer in Physics. Then, create a new DataFrame that displays each subject and the number of students who scored above 90 in that subject."


````python
student_data = pd.DataFrame({
    'Student_ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Math': [92, 85, 88, 76, 95],
    'Physics': [85, 90, 82, 89, 92],
    'Chemistry': [88, 85, 91, 87, 93],
    'Biology': [90, 88, 94, 86, 89]
})
````

````python
student_data
````

````python
#The average score for each student across all subjects
Average_score=pd.concat((student_data['Name'],student_data.iloc[:,2:].mean(axis=1)),axis=1).rename(columns={0:'average_score'})

#The top scorer in Physics
top_scorer=student_data.loc[student_data['Physics'].idxmax(),'Name']

#Subject and the number of students who scored above 90 in that subject
Score_above_90=(student_data[['Math','Physics','Chemistry','Biology']]>90).sum()

print(f'average score for each student across all subjects:')
print(Average_score)
print(f'\nThe top scorer in Physics: {top_scorer}')
print(f'\nSubject and the number of students who scored above 90 in that subject')
print(Score_above_90)
````
